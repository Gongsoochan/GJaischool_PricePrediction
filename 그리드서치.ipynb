```python
# x데이터, y데이터 분류
y = df1[df1['지역']=='서울']['시금치가격']
x = df1[df1['지역']=='서울'].drop(['시금치가격','날짜','생산자물가','소비자물가','가격','지역'], axis = 1)

# train데이터, test데이터 분류
tr_x, tt_x, tr_y, tt_y = train_test_split(x, y, random_state = 3, test_size = 0.2)
tr_x.shape, tt_x.shape, tr_y.shape, tt_y.shape

# train데이터를 기준으로 minmax스케일러 적용
mms = MinMaxScaler().fit(tr_x)
mms_train_x = mms.transform(tr_x)
mms_test_x = mms.transform(tt_x)
```

## 모델 테스트

```python
# 리니어모델 테스트

linear_m = LinearRegression()
linear_m.fit(mms_train_x, tr_y)

print(cross_val_score(linear_m, mms_train_x, tr_y).mean())
print(linear_m.score(mms_train_x, tr_y), linear_m.score(mms_test_x, tt_y))
```

```python
# 디시젼트리모델 테스트
decision_m = DecisionTreeRegressor()
decision_m.fit(mms_train_x, tr_y)

print(cross_val_score(decision_m, mms_train_x, tr_y).mean())
print(decision_m.score(mms_train_x, tr_y), decision_m.score(mms_test_x, tt_y))
```

```python
# 1번
#q = PolynomialFeatures(degree = 5)
#x_q = q.fit_transform(mms_train_x)
#x_tt_q = q.transform(mms_test_x)

m = Lasso()
m.fit(mms_train_x, tr_y)
print(cross_val_score(m, mms_train_x, tr_y).mean())
print(m.score(mms_train_x, tr_y), m.score(mms_test_x, tt_y))


m1 = Ridge()
m1.fit(mms_train_x, tr_y)
print(cross_val_score(m1, mms_train_x, tr_y).mean())
print(m1.score(mms_train_x, tr_y), m1.score(mms_test_x, tt_y))


m2 = SGDRegressor(max_iter=100, shuffle=False)#, verbose=1)#, eta0=1e-4)
m2.fit(mms_train_x, tr_y)
print(cross_val_score(m2, mms_train_x, tr_y).mean())
print(m2.score(mms_train_x, tr_y), m2.score(mms_test_x, tt_y))
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

tr_x, tt_x, tr_y, tt_y = train_test_split(x, y, test_size=0.2, random_state=42)

#그리드서치 파라미터 키값으로 모델 파라미터 이름, 밸류값으로 파라미터에 넣을 값 리스트로
param = {
    'n_estimators': [50,100,150],
    'max_features':[6,7,8,9,10,15,20,len(x.columns)]
}

#스케일링
mms = MinMaxScaler().fit(tr_x)
mms_train_x = mms.transform(tr_x)
mms_test_x = mms.transform(tt_x)

#랜덤포레스트 모델 생성 및 학습
rf = RandomForestRegressor(random_state=42)
rf.fit(mms_train_x,tr_y)
print("랜덤포레스트 피처중요도", rf.feature_importances_) #중요도 총합 1로 오름차순 출력

#그리드서치 하이퍼파라미터 찾기
grid_search = GridSearchCV(estimator=rf, param_grid=param, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(mms_train_x, tr_y)

#그리드서치 결과물 출력 파라미터, 점수 등
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
print("Best_estimator", grid_search.best_estimator_)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV

# 데이터 로드
y = df1[df1['지역']=='서울']['시금치가격'].values
x = df1[df1['지역']=='서울'].drop(['시금치가격','생산자물가','날짜','가격','지역'], axis = 1)

tr_x, tt_x, tr_y, tt_y = train_test_split(x, y, test_size=0.2, random_state=42)

param = {
    'n_estimators': [30,50,100,150],
    'learning_rate': [0.01, 0.1, 0.2],
}

mms = MinMaxScaler().fit(tr_x)
mms_train_x = mms.transform(tr_x)
mms_test_x = mms.transform(tt_x)

# 그래디언트 부스팅 모델 생성 및 학습
gb = GradientBoostingRegressor(random_state=42)
grid_search = GridSearchCV(estimator=gb, param_grid=param, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(mms_train_x, tr_y)
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import GridSearchCV

# 데이터 로드
y = df1[df1['지역']=='서울']['시금치가격'].values
x = df1[df1['지역']=='서울'].drop(['시금치가격','생산자물가','날짜','가격','지역'], axis = 1)

tr_x, tt_x, tr_y, tt_y = train_test_split(x, y, test_size=0.2, random_state=42)

param = {
    'n_estimators': [30, 50, 100, 150,200,300],
    'learning_rate': [0.01, 0.1, 0.2],
}

mms = MinMaxScaler().fit(tr_x)
mms_train_x = mms.transform(tr_x)
mms_test_x = mms.transform(tt_x)

# 그래디언트 부스팅 모델 생성 및 학습
ab = AdaBoostRegressor(random_state=42)
grid_search = GridSearchCV(estimator=ab, param_grid=param, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(mms_train_x, tr_y)
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
```
