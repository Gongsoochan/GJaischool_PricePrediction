```python
import pandas as pd
import numpy as np

df = pd.read_csv('토마토_전지역.csv', encoding = 'cp949')
df.drop(['True', 'Unnamed: 0'], axis = 1, inplace = True)

g_df = df[df['지역'] == '광주']
g_df.sort_values('날짜', inplace = True, ignore_index=True)
# g_df.reset_index(drop = True, inplace = True)
```

```python
rain=g_df[['지역','연','1시간 최다강수량(mm)', '일강수량(mm)']]
rain_hr=[]
for i in range(len(rain)):
    if int(rain['일강수량(mm)'].iloc[i])>20:
        if int(rain['1시간 최다강수량(mm)'].iloc[i])>3:
            
            rain_hr.append(1)
        else:
            rain_hr.append(0)
    else:
        rain_hr.append(0)
            
g_df['강수량ox'] = rain_hr
```

```python
g_df['추가'] = g_df['날짜']
g_df
```

```python
def fix(x):
    y, m, d = x.split('-')
    return int(m) * 100 + int(d)
g_df['추가'] = g_df['추가'].apply(fix)
```

```python
import  datetime

len_gangsu = []
len_temp_max = []
len_temp_min = []
j_len = []
pok_list = []
neng_list = []
for i in range(len(g_df)):
    y, m, d = g_df['날짜'].iloc[i].split('-')
    base_date = datetime.date(int(y), int(m), int(d))
    count = 0
    max_avg = 0
    min_avg = 0
    j_avg = 0
    pok_count = 0
    neng_count = 0
    
    for i in range(22, 53):
        base_date = base_date + datetime.timedelta(days = -i)
        input_date = base_date.strftime('%Y-%m-%d')

        chuga = g_df[g_df['날짜'] == input_date]
        if 0 == len(chuga):
            pass
        else:
            max_temp = int(chuga['최고기온(°C)'].values)
            min_temp = int(chuga['최저기온(°C)'].values)
            j_value = int(chuga['평균 증기압(hPa)'].values)
            
            max_avg += max_temp
            min_avg += min_temp
            j_avg += j_value
            
#             if i > 10:
            if chuga['강수량ox'].values == 1:
                count += 1000
                
#             if i < 31: 
            if max_temp > 33:
                pok_count += 1

            if min_temp <= 4 and max_temp >= 16:
                neng_count += 1

    len_gangsu.append(count)
    pok_list.append(pok_count)
    neng_list.append(neng_count)
    j_len.append(round(j_avg / 30, 1))
    len_temp_max.append(round(max_avg / 30, 1))
    len_temp_min.append(round(min_avg / 30, 1))
    
g_df['30일강우횟수'] = len_gangsu
g_df['30일평균최고기온'] = len_temp_max
g_df['30일평균최저기온'] = len_temp_min
g_df['30일평균증기압'] = j_len
g_df['30일폭염횟수'] = pok_list
g_df['30일일교차큰날횟수'] = neng_list
g_df
```
