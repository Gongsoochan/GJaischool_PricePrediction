```python
import pandas as pd
import matplotlib.pyplot as plt
import re


# 데이터프레임 로드
df = pd.read_csv('일일기상데이터.csv', encoding = 'cp949')
df
```

```python
# 모든 컬럼 결측치 확인하기

for idx, column in enumerate(df.columns):
    print(f'{idx + 1}. {column} : {df[column].isnull().sum()}개')
```

```python
import re

# 시각이 포함된 컬럼 추출
drop_column_list = []

for column in df.columns:
    if re.search('시각', column):
        drop_column_list.append(column)
        
# 시각이 포함된 컬럼과 필요없는 컬럼 삭제
d_df = df.drop(drop_column_list, axis = 1)
d_df
```

```python
# 강수, 적설은 비, 눈이 오지 않는 경우 결측치로 채워지는 것으로 파악됨
# 강수 시간 또한 강수가 없을 경우 0이 됨으로 0으로 채움
# 그래서 결측치를 0으로 채워준다.

import re

#  강수, 적설 포함된 컬럼 추출
drop_column_list = []

for column in d_df.columns:
    if re.search('강수', column) or re.search('적설', column):
        drop_column_list.append(column)

# 시간을 제외한 나머지 컬럼의 결측치를 0으로 채움
for column in drop_column_list:
    d_df[column].fillna(0, inplace = True)
    
d_df
```

```python
# 강수, 적설을 0으로 결측치를 채운 후 다른 결측치 확인

for idx, column in enumerate(d_df.columns):
    print(f'{idx + 1}. {column} : {d_df[column].isnull().sum()}개')
```

```python
# 결측치가 50퍼센트가 넘는 컬럼 결측치 확인하기

nan_50_list = []

for idx, column in enumerate(d_df.columns):
    if (d_df[column].isnull().sum() / len(d_df)) >= 0.5:
        print(f'{idx + 1}. {column} : {d_df[column].isnull().sum()}개')
        nan_50_list.append(column)
```

```python
# 결측치가 50퍼센트가 넘는 컬럼 삭제하기

d_df.drop(nan_50_list, axis = 1, inplace = True)
d_df
```

```python
# 결측치 확인

for idx, column in enumerate(d_df.columns):
    print(f'{idx + 1}. {column} : {d_df[column].isnull().sum()}개')
```

```python
# 풍향, 운량 컬럼 삭제

fid = []

for column in d_df.columns:
    if re.search('풍향', column) or re.search('운량', column):
        fid.append(column)
    
# + 지점, 풍접합도 드랍
d_df.drop(fid + ['지점', '풍정합(100m)'], axis = 1, inplace = True)
d_df
```

```python
# 결측치 확인

for idx, column in enumerate(d_df.columns):
    print(f'{idx + 1}. {column} : {d_df[column].isnull().sum()}개')
```

```python
import numpy as np

# 평균기온, 최저기온, 최고기온의 결측치를 전날, 다음날의 평균기온의 평균으로 결측치 채움


for columns in ['평균기온(°C)', '최저기온(°C)', '최고기온(°C)']:
    idx_list = d_df[d_df[columns].isnull()].index

    for idx in idx_list:
        if idx == 0:
            i = 1
            while True:
                    if str(d_df[columns][idx + 1]) != 'nan':
                        d_df[columns][idx] = d_df[columns][idx + i]
                        break
                    else:
                        i += 1
        elif idx == (len(d_df) - 1):
            d_df[columns][idx] = d_df[columns][idx - 1]
        else:
            i = 1
            while True:
                    if str(d_df[columns][idx + i]) != 'nan':

                        d_df[columns][idx] = round((d_df[columns][idx - 1] + d_df[columns][idx + i]) / 2, 2)
                        break
                    else:
                        i += 1
                    
                    
                    
    
```

```python
# 연, 월, 일 컬럼 추가
# 일시 컬럼은 merge이후에 삭제

year_list = []
month_list = []
date_list = []

for idx in range(len(d_df)):
    year, month, date = d_df['일시'][idx].split('-')
    year_list.append(int(year))
    month_list.append(int(month))
    date_list.append(int(date))

print(len(year_list), len(month_list), len(date_list), len(d_df))

values_list = [year_list, month_list, date_list]

for idx, column in enumerate(['연', '월', '일']):
    d_df[column] = values_list[idx]
```

```python
count = 0
l = []

for idx, gang_su in enumerate(d_df['일강수량(mm)'].values):
    if idx == 0:
        if gang_su != 0:
            l.append(1)
        else:
            l.append(0)
        continue
        
    if gang_su != 0:
        l.append(1 + l[idx - 1])
    else:
        l.append(0)
```
